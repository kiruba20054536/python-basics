## EXERCISE

1. Find how many odd and even numbers are there in betweenn 2 numbers.

```python
a=int(input("Enter a number:"))
b=int(input("Enter another number:"))

count1=0
count2=0
for i in range(a,b+1):
    if i%2==0:
        #print(i) # to print the even numbers
        count1+=1
    elif i%2!=0:
        #print(i)  # to print the odd numbers
        count2+=1
print("Number of even:",count1)
print("Number of odd:",count2)
```

2. count the number which are divisible by 3&5, Range 1-100

```python
count=0
for i in range(1,101):
    if i%3 == 0 and i%5==0:
        count+=1
print(count)
```

4. get input of 10 numbers and find the average for that numbers.

```python
a = []
for i in range(1, 11):
    b = int(input("Enter number " + str(i) + ": "))
    a.append(b)

length=len(a)
total = 0
for i in a:
    total += i
print("Average of numbers:",total/length)
```

5. write a code to display n terms of natural numbers along with their total.

```python
total =0
n= int(input("Enter a number:"))
for i in range(1,n+1):
    print(i,end=",")
    total +=i
print("\nTotal Value is:",total)
    
```
