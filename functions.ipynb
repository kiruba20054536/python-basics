3) Get a integer number from user and pass it to the function called findpassorfail(). Let the function print "pass" if the mark is above 35 and fail if the mark is below 35.

```python
b=int(input("Enter your Mark:"))
def findPassOrFail(b):
    if b >= 35:
        print("You're PASS")
    else:
        print("You're Fail")
findPassOrFail(b)
```

4) Get input for a and b and pass it to the function called printrange() let the function print numbers from a to b.

```python
a=int(input("Enter a number:"))
b=int(input("Enter another number:"))
def printrange(a,b):
    print("In-Between Numbers")
    for i in range(a,b+1):
        print(i)
printrange(a,b)
```
