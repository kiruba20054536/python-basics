4. Make a calculator

```python
a = int(input("A:"))
b = int(input("B:"))
Operation = input("ADD+,SUB-,DIVIDE/,MULTI*:")
if Operation == "+":
    print(a+b)
elif Operation == "-":
    print(a-b)
elif Operation == "*":
    print(a*b)
elif Operation == "/":
    print(a/b)
else:
    print("INVALID OPERATOR")
```

7. Checks whether student requires additional class or not based on average mark.

```python
a= int(input("sub 1 mark:"))
b= int(input("sub 2 mark:"))
c= int(input("sub 3 mark:"))
d= int(input("sub 4 mark:"))
e= int(input("sub 5 mark:"))
avg=(a+b+c+d+e)/5
if avg < 35:
    print("Additional class required")
else:
    print("You are good to go")
```
