# Continue

```python
for i in range(1,10):
    if i==3:
        continue # "Continue" statement will skip the iteration when i=3 and continue the rest of the loop
    print(i)
```

# Break

```python
for i in range(1,11):
    if i==5:
        break  # "break" statement will terminate the entire loop when the condition is satisfied
    print(i)
```

# For-loop Else

```python
for i in range (0,10):
    if i==5:
        break
    print(i)
else:   # else clause will execute only when the loop fully runs without break
    print("Else plot is executed") # here break runs when i=5 i.e) else doesn't runs and breaks after prints 4
```

# while-loop else

```python
i=10
while i<16:
    if i==14:
        break
    print(i)
    i+=1
else:
    print("Else plot is executed")
```
