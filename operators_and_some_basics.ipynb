```python
a = 10
b = 10 
c = a
d=[10,20,30]
e=[10,20,30]

print(a is not b)
print(a is c)
print(id(d)) # to check Memory address
print(id(e))
print(d is e)
print(d is not e)
print(d==e)
# even though the elements are same in the list d and e both stored in different memory address so it returns False 
# but while using equal == operator it returns True because == opertor Checks values inside the objects, Compares elements one by one
```

## Membership Operators

```python
x = 24
y = 20
list = [10, 20, 30, 40, 50]

if (x not in list):
    print("x is NOT present in given list")
else:
    print("x is present in given list")

if (y in list):
    print("y is present in given list")
else:
    print("y is NOT present in given list")
```
