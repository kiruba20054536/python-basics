```python
n=5 # no.of rows
for i in range(1,n+1):
    for j in range(1,i+1):
        print("*",end="")
    print() 
```

```python
n=5
for i in range(n,0,-1):
    for j in range(1,i+1):
        print("*",end="")
    print()
```

```python
for i in range(1,6):
    for j in range(1,i+1):
        print(j,end="")
    print()
```

```python
n=4 # no.of rows
k=1 # temporary variable
for i in range(1,n+1):
    for j in range(1,k+1):
        print("*",end="")
    k+=2
    print()
```
