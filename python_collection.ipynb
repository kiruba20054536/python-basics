# LIST: []
### Allows Duplicate.
### Any type of data can be stored.
### We can modify the list item.
### We can add or remove.

```python
a=[1,2,3,4]
print(a)
print(a[1]) # INDEXING
```

```python
# insert(), append(), extend(), pop()
```

```python
#append adds value at the last position of the variable
a.append([7,70]) # append takes only one argument
print(a) 
```

```python
a.append('kiruba')
print(a)
```

```python
# add the value 12 on the 1st index position
a.insert(1,12) # insert(position, value)
print(a)
```

```python
# remove 1 and change into 11
a[0]=11
print(a)
```

```python
# to remove 4
a.pop(4) # here 4 refers to the index position
print(a)
```

```python
# to remove the last element, just run pop() without an argument.
a.pop()
print(a)
```

```python
b=[11,23,12]
a.extend(b) # to add two list.
print(a)
```

## TUPLE: ()
### Allows Duplicate.
### Any type of data can be stored.
### We cannot modify the tuple item.
### we cannot add or remove.

```python
c=(5,6,7,8)
print(c)
```

```python
# casting - change tuple to list
d=list(c)
print(d)
```

## SET: {}
### Do not allow Duplicate, Duplicate values will be removed.
### Any type of data can be stored
### We cannot modify the set item but we can add or remove items.
### Sets are unordered.
### add(),update(), remove(),pop()

```python
e={1,2,"kiruba",3,4,1} # duplicate removed
print(e)
```

```python
e.add(8)
print(e)
```

```python
e.remove(1) # 1 refers to the value not index
print(e)
```

```python
e.pop() # removes 1st element
print(e)
```

## DICTIONARY: {}
### Do not allow Duplicate, Duplicate value will overwrite existing value.
### Any type of data can be stored.

```python
f= {
    "Name":"Kiruba",
    "Age":21,
    "Location":"Kovilpatti",
    "Students":["bala","john","yuva"]
}
print(f)
print(f["Location"])
```

```python
print(f.keys())
```

```python
print(f.values())
```

```python
# Modifying or updating
f["Age"]=22
# OR 
# f.update({"Age":22})
print(f)
```

```python
#Adding new data
f["college"]="MCC"
print(f)
```

```python
# to delete a value
f.pop("Students")
# OR
# del f["Students"]
print(f)
```

```python
# to delete a complete dictionary
del f
```

```python
print(f)
```
